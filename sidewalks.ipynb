# Walkability Data Exploration
This notebook explores the `npt-sidewalks` dataset using the geopandas library.

```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
file = "./datasets/sidewalk-map/npt-sidewalks.geojson"
dataset = gpd.read_file(file)
```

## Plot the dataset

```python
dataset.plot(figsize=(8, 12))
plt.show()
```

## Crosswalks
Some of the shapes in the sidewalk dataset are tagged as crosswalks.

```python
crosswalks = dataset[dataset['surface']=='crosswalk']
crosswalks.centroid.plot(figsize=(8,12))
plt.show()
```

We can extract more crosswalks by looking at the intersection of the sidewalk map with road centerlines (performed using QGIS).

```python
crossings_file = "./datasets/sidewalk-map/npt-sidewalk-street-intersections.geojson"
crossings = gpd.read_file(crossings_file)
crossings.plot(figsize=[8,12])
plt.show()
```
